```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.plyplot as plt
import matplotlib.ticker as PercentFormatter
```

```python
import pandas as pd
df = pd.read_excel("/content/DATA_Redes Sociales.xlsx")
df
```

```python
print("Los tipos de datos son:")
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
from matplotlib import pyplot as plt
df['Likes'].plot(kind='hist', bins=40, title='Frecuencia de Likes')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
import plotly.express as px
fig = px.bar(df, x='Likes')
fig.show()
```

```python
fig = px.pie(df, values='Likes', names='Nombre_Red')
fig.update_traces(marker = dict(line = dict(color = 'black', width = 2)))
fig.show()
```

```python
fig = px.pie(df, values='Likes', names='Vistas', color_discrete_sequence=px.colors.sequential.RdBu, hole = 0.5)
fig.update_traces(marker = dict(line = dict(color = 'black', width = 2)))
fig.show()
```

```python
fig = px.scatter(df, x="Likes", y="Vistas", size='Likes')
fig.update_traces(marker = dict(line = dict(color = 'black', width = 2)))
fig.show()
```

```python
fig = px.box(df, x="Nombre_Red", y="Likes")
fig.show()
```
