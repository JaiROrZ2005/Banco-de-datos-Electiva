```python
from matplotlib import pyplot as plt
df['Edad '].plot(kind='hist', bins=20, title='Edad ')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
df['Notas'].plot(kind='hist', bins=40, title='Notas')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df["Sexo_Estudiante"].value_counts()
```

**Grafico de Barras**

En este diagrama vemos la comparación de la cantidad de estudiantes de cada sexo
donde evidenciamos que el sexo masculino tiene un total de 30 estudiantes
y el sexo femenino tiene un total de 13 estudiantes

```python
import plotly.express as px
fig = px.bar(df, x='Sexo_Estudiante')
fig.show()
```

**Grafico de Pastel**

Compara el porcentaje de los diferentes sexos encuestados

```python
fig = px.pie(df, values='Notas', names='Sexo_Estudiante')
fig.update_traces(marker = dict(line = dict(color = 'black', width = 2)))
fig.show()
```

**Grafico de Pastel (Estilo Dona)**

Este grafico compara los porcentajes de cada nota que los estudiantes han obtenido

```python
fig = px.pie(df, values='Notas', names='Notas', color_discrete_sequence=px.colors.sequential.RdBu, hole = 0.5)
fig.update_traces(marker = dict(line = dict(color = 'black', width = 2)))
fig.show()
```

```python
fig = px.line(df, x="Notas", color='Sexo_Estudiante')
fig.show()
```

**Diagrama de Dispersión**

En este evidenciamos la dispersión de las notas comparando los sexos donde vemos que la nota más alta la tiene el sexo masculino y la más baja el sexo femenino.

```python
fig = px.scatter(df, x="Notas", y="Sexo_Estudiante", size='Notas')
fig.update_traces(marker = dict(line = dict(color = 'black', width = 2)))
fig.show()
```

**Grafico de Caja**

En este grafico representamos la media, los cuartiles, los maximos y los minimos de cada sexo.

```python
fig = px.box(df, x="Sexo_Estudiante", y="Notas")
fig.show()
```
